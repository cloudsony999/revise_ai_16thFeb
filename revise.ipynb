resizi the image


```python
size_small=(600,600)
size_tiny=(200,200)
```

```python
for f in os.listdir('.'):
    if f.endswith(".jpg"):
        im=i.open(f)
        fn,stext=os.path.splitext(f)
        im.thumbnail(size_small)
        im.save("saveimg/{}_small.{}".format(fn,stext))
        im.thumbnail(size_tiny)
        im.save("saveimg/{}_tiny.{}".format(fn,stext))
        
```
